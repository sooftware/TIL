```python
for each_tag in soup.find_all('p'):
    print(each_tag.get_text())
```

```python
links = soup.find_all('a')
links
```

```python
for each in links:
    href = each['href']
    text = each.string
    print(text + ' -> ' + href)
```

거기서 **href** 속성을 찾으면 링크 주소를 얻을 수 있다

## 3-3 실전: 시카고 샌드위치 맛집 소개 사이트에 접근하기

```python
print(soup.find_all('div','sammy'))
```

## 3-4 접근한 웹 페이지에서 원하는 데이터 추출하고 정리하기

그럼 이제 div의 sammy 태그에서 우리가 원하는 정보를 얻어보자

```python
tmp_one = soup.find_all('div','sammy')[0]
type(tmp_one)
```

우리가 찾은 정보는 bs4.element.Tag라는 클래스의 객체인가보다  

```python
tmp_one.find(class_='sammyRank')
```

그럼 이 tmp_one에서 'sammyRank'를 찾아보자  
그리고 이 결과에 대해서 텍스트를 취해보자

```python
tmp_one.find(class_='sammyRank').get_text()
```

위와 똑같은 방법으로 리스트도 얻어보자

```python
tmp_one.find(class_='sammyListing').get_text()
```

메뉴이름과 가게이름이 같이 나오긴 했지만 원하는 정보를 얻었다

```python
tmp_one.find('a')['href']
```

또한 a 태그에서 href 정보를 가지고 클릭했을 때 연결될 주소도 저장할 수 있다  
그리고 위의 sammyListing의 경우 메뉴이름과 가게 이름이 같이 있어서 분리해야 한다  
이 분리는 re 라이브러리의 split 명령을 통해 얻어오자  
물론 그냥 문자열 split을 해도 된다  
일단 책이 re 의 split을 쓰니 그대로 쓴다

```python
import re

tmp_string = tmp_one.find(class_='sammyListing').get_text()

re.split(('\n|\r\n'), tmp_string)

print(re.split(('\n|\r\n'), tmp_string)[0])
print(re.split(('\n|\r\n'), tmp_string)[1])
```

re는 Regular Express 즉 정규식의 줄임말이다  
re.split는 내가 지정한 특정 패턴이일치하면 분리시킨다  
위의 코드는 '\n'이나 '\r\n'이 오게 되면 분리시키도록 지정했다

이제 이 방식대로 50개에 대해서 for 문만 돌리면 되겠구나!! 싶었지만 하나 문제점이 있다  
50개의 Rank 리스트 중 어떤 경우는 상대경로로, 어떤 경우는 절대경로로 나오기 때문이다  
이럴 때 사용하는 것이 urllib에 있는urljoin이라는 명령이다   
  
이 명령을 사용하면 절대경로로 잡힌 url은 그대로 두고 상대경로로 잡힌 url은 절대경로로 변경할 수 있다

```python
from urllib.parse import urljoin

rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div','sammy')

for item in list_soup:
    rank.append(item.find(class_='sammyRank').get_text())
    tmp_string = item.find(class_='sammyListing').get_text()
    
    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0])
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])
    
    url_add.append(urljoin(url_base, item.find('a')['href']))
```

위는 앞에서 배운 내용을 종합적으로 정리해서 만든 코드이다  
내가 필요한 정보를 전부 저장했다  
그리고 잘 저장이 되었는지를 확인하기 위해 아래의 head 함수를 정의했다  
head 함수는 pandas 처럼 개의 요소만 확인하려고 만들어본 함수이다   
  
  
※ 주의사항 ※  
  
re.split() 사용시 습관적으로 '\n | \r\n' 와 같이 or 명령 중간에 공백을 넣었는데  
re.split() 시에는 이것도 문자로 인식해서 의도와 다른 방향으로 코드가 진행된다

```python
def head(list_):
    print(list_[:5])
```

```python
head(rank)
```

```python
head(main_menu)
```

```python
head(cafe_name)
```

```python
head(url_add)
```

```python
print(len(rank))
print(len(main_menu))
print(len(cafe_name))
print(len(url_add))
```
